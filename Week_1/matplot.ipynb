```python
import numpy as np
import matplotlib.pyplot as plt


normal_data = np.random.normal(loc=0, scale=1, size=1000)
beta_data = np.random.beta(a=2, b=5, size=1000)
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].hist(beta_data, bins=20, color='skyblue', edgecolor='black')
axes[0].set_title('Beta Distribution (a=2, b=5)')


axes[1].hist(normal_data, bins=20, color='salmon', edgecolor='black')
axes[1].set_title('Normal Distribution (Mean=0, Std=1)')
plt.tight_layout()
plt.show()
```

```python
data = np.random.normal(loc=0, scale=1, size=1000)
mean_val = np.mean(data)
p25 = np.percentile(data, 25)
p75 = np.percentile(data, 75)


plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, color='lightpink', edgecolor='black', alpha=0.8, label='Frequency')


plt.axvline(mean_val, color='crimson', linestyle='--', linewidth=2, label=f'mean ({mean_val})')
plt.axvline(p25, color='blue', linestyle=':', linewidth=2, label=f'25th % ({p25})')
plt.axvline(p75, color='green', linestyle=':', linewidth=2, label=f'75th % ({p75})')


y_top = plt.gca().get_ylim()[1]
plt.text(mean_val, y_top*0.9, 'Mean', color='crimson', ha='center', fontweight='bold')
plt.text(p25, y_top*0.8, '25th %', color='blue', ha='center', fontweight='bold')
plt.text(p75, y_top*0.8, '75th %', color='green', ha='center', fontweight='bold')


plt.title('Histogram of Normal Distribution', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)

# Show the plot
plt.show()
```
