```python
import numpy as np
arr=np.random.rand(3,6)
print(arr)
```

```python
arr1=arr+[0, 0, 2, 4, 5, 3]
print(arr1)
```

```python
arr2=arr1.reshape(9,2)
print(arr2)
```

```python
arr3=arr2.T
print(arr3)
```

```python
mean=np.mean(arr3)
for i in range (2) :
    for j in range (9):
     if arr3[i,j]>mean:
        print ([i,j])

```
